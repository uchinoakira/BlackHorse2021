### 购物篮分析
数据集：MarketBasket

下载地址：https://www.kaggle.com/dragonheir/basket-optimisation

该数据集为rawdata，没有记录TransactionID和列名

ToDo：统计交易中的频繁项集和关联规则

```python
from efficient_apriori import apriori
import pandas as pd
```

```python
data = pd.read_csv('./Market_Basket_Optimisation.csv', header = None)
data.head()
```

```python
transactions = []
for i in range(0,data.shape[0]):
    temp = []
    for j in range(0,data.shape[1]):
        if str(data.values[i,j]) != 'nan':
            temp.append(data.values[i,j])
    transactions.append(temp)
```

```python
itemsets,rules = apriori(transactions,min_support = 0.05, min_confidence = 0.2)
print("频繁项集：",itemsets)
print("关联规则：",rules)
```
